```python
from scipy.spatial.transform import Rotation as R

import numpy as np
import matplotlib.pyplot as plt
```

```python
r = R.from_euler('xyz', [10, 10, 0], degrees=True)
```

```python
rot_vec = r.as_rotvec()
```

```python
ax = plt.figure().add_subplot(projection='3d')
# Labelling X-Axis
ax.set_xlabel('X-Axis')
 
# Labelling Y-Axis
ax.set_ylabel('Y-Axis')
 
# Labelling Z-Axis
ax.set_zlabel('Z-Axis')
ax.quiver(0, 0, 0, *rot_vec, length=0.1, normalize=True)
plt.show()
```
